```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('train.csv')
df.columns
```

```python
dataset = df.drop(['Name','Cabin','Ticket','PassengerId'], axis=1)
```

```python
dataset.isna().sum()
```

```python
le = preprocessing.LabelEncoder()
```

```python
dataset['Sex'] = le.fit_transform(dataset['Sex'])
```

```python
dataset['Embarked'] = le.fit_transform(dataset['Embarked'])
```

```python
from sklearn import neighbors
```

```python
y = dataset['Pclass']
x= dataset.drop(['Pclass'], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30, random_state = 0)
```

```python
def machine_knn(x_train, x_test, y_train, y_test):
    print('This model is for Pclass:/n')
    for i in range(1, x_test.shape[0]+1):
        knn = neighbors.KNeighborsClassifier(n_neighbors = i)
        print('The score value with k=',i)
        print(knn.fit(x_train, y_train).score(x_test,y_test))
        print('The Confusion Matrix value for k=',i)
        y_pred = knn.predict(x_test)
        print(confusion_matrix(y_test, y_pred),'\n')
    return y_pred
```

```python
machine_knn(x_train, x_test, y_train, y_test)
```
